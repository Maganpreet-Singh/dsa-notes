# Brute Force TLE Error

```python
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

for num in m:
    count = 0
    for x in n:
        if x == num:
            count += 1
    print(f"{num}: {count}")
```

# Optimal solution

```python
hash_list = [0] * 11
for num in n:
    hash_list[num] += 1

for num in m:
    if num < 1 or num > 10:
        print(f"{num}: 0")
    else:
        print(f"{num}: {hash_list[num]}")
```

##### Time Complexity = O(n+m)

##### Space Complexity = O(1)

# Dictionary Method

```python
freq = {}
for num in n:
    freq[num] = freq.get(num,0) +1
for num in m:
    if num in freq:
        print(f"{num}:{freq[num]}")
    else:
        print(f"{num}:0")
```

```python
s = "azyxyyzaaaa"
q = ["a", "a", "y", "u"]
hash_list = [0] * 26
for ch in s:
    index = ord(ch) - ord('a')
    hash_list[index] += 1

for ch in q:
    index = ord(ch) - ord('a')
    print(f"{ch} -> {hash_list[index]}")
```
